```python
normal_day = ['Sunday']
normal_start_times = ['12', '1', '2', '3', '4', '5','6']
```

```python
def abnormal(a, b):
    normal_day = 'Sunday'
    normal_start_times = ['12', '1', '2', '3', '4', '5','6']
    if a == normal_day and b in normal_start_times:
        return 0
    else:
        return 1
    
```

```python
df['abnormal_start'] = df.apply(lambda row: abnormal(row['day_of_week'], row['start_time']), axis=1)

```

<b>Wins</b> 

```python
win_reg = r'^[0-9]{1,2}'
```

```python
df.home_wins[0]
```

```python
re.findall(win_reg, df.home_wins[0])[0]
```

```python
def normalize_wins(a):
    win_reg = r'^[0-9]{1,2}'
    a = re.findall(win_reg, a)[0]
    return a

    
```

```python
df['home_wins'] = df.apply(lambda x: normalize_wins(x['home_wins']), axis=1)
```

```python
df['away_wins'] = df.apply(lambda x: normalize_wins(x['away_wins']), axis=1)
```

```python
df['playoff_implications'] = 0
```

```python
def playoff_eval(df):
    for index, row in df.iterrows():
        if df.week_num[index] >= 10:
            cut_off = np.floor(df.week_num[index]/2) - 1
            if df.home_wins[index] >= cut_off or df.away_wins[index] >= cut_off:
                df.playoff_implications[index] = 1
    
```

```python
playoff_eval(df)
```
