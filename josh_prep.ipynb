```python
df.date = pd.to_datetime(df.date)
```

```python
time_reg = r'^[1-9]{1,2}'
```

```python
df['start_time'][0]
```

```python
df['start_time'] = df['start_time'].apply(lambda x: re.findall(time_reg, str(x))[0])
```

```python
df
```

```python
normal_day = ['Sunday']
normal_start_times = ['12', '1', '2', '3', '4', '5','6']
```

```python
def abnormal(a, b):
    normal_day = 'Sunday'
    normal_start_times = ['12', '1', '2', '3', '4', '5','6']
    if a == normal_day and b in normal_start_times:
        return 0
    else:
        return 1
    
```

```python
df['abnormal_start'] = df.apply(lambda row: abnormal(row['day_of_week'], row['start_time']), axis=1)

```

<b>Wins</b> 

```python
win_reg = r'^[0-9]{1,2}'
```

```python
df.home_wins[0]
```

```python
re.findall(win_reg, df.home_wins[0])[0]
```

```python
def normalize_wins(a):
    win_reg = r'^[0-9]{1,2}'
    a = re.findall(win_reg, a)[0]
    return a

    
```

```python
df['home_wins'] = df.apply(lambda x: normalize_wins(x['home_wins']), axis=1)
```

```python
df['away_wins'] = df.apply(lambda x: normalize_wins(x['away_wins']), axis=1)
```
